```python
import pandas as pd

# Read the CSV file into a pandas dataframe
data = pd.read_csv("sales_data.csv")

# Display the first 5 rows of the dataframe
print("First 5 rows of data:")
print(data.head())

# Display the shape of the dataframe (number of rows, number of columns)
print("\nDataframe shape:")
print(data.shape)

# Display the total sales for each product
total_sales = data.groupby('Product')['Sales'].sum()
print("\nTotal sales for each product:")
print(total_sales)

# Display the average sales per month
monthly_sales = data.groupby('Month')['Sales'].mean()
print("\nAverage sales per month:")
print(monthly_sales)

# Display the top 5 salespeople by total sales
top_salespeople = data.groupby('Salesperson')['Sales'].sum().nlargest(5)
print("\nTop 5 salespeople by total sales:")
print(top_salespeople)
```

```python
import pandas as pd

# Read the CSV file into a pandas dataframe
data = pd.read_csv("data.csv")

# Data cleaning
# Remove any rows with missing values
data = data.dropna()

# Data exploration
# Display the unique values in the "Category" column
print("Unique Categories:")
print(data["Category"].unique())

# Display the distribution of values in the "Category" column
print("\nCategory Value Counts:")
print(data["Category"].value_counts())

# Data manipulation
# Add a new column that calculates the total cost of each item
data["Total Cost"] = data["Price"] * data["Quantity"]

# Analyzing the data
# Calculate the average and total sales for each category
category_sales = data.groupby("Category").agg({"Total Cost": ["mean", "sum"]})
print("\nCategory Sales:")
print(category_sales)
```

```python
import pandas as pd

# Read the stock data CSV file into a pandas dataframe
stock_data = pd.read_csv("stock_data.csv")

# Read the bond data CSV file into a pandas dataframe
bond_data = pd.read_csv("bond_data.csv")

# Calculate the average returns for each asset
stock_average_return = stock_data["Return"].mean()
bond_average_return = bond_data["Return"].mean()

# Calculate the standard deviation of returns for each asset
stock_return_std = stock_data["Return"].std()
bond_return_std = bond_data["Return"].std()

# Calculate the correlation coefficient between the two assets
correlation_coefficient = stock_data["Return"].corr(bond_data["Return"])

# Print out the results
print("Stock Average Return:", stock_average_return)
print("Bond Average Return:", bond_average_return)
print("Stock Return Standard Deviation:", stock_return_std)
print("Bond Return Standard Deviation:", bond_return_std)
print("Correlation Coefficient:", correlation_coefficient)
```
